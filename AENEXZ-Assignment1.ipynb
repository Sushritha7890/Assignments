```python
#1. Write a program to find the largest of three numbers.

def find_largest(num1, num2, num3):
   
    if (num1 >= num2) and (num1 >= num3):
        largest = num1
    elif (num2 >= num1) and (num2 >= num3):
        largest = num2
    else:
        largest = num3

    return largest
    
a = 15
b = 42
c = 8

largest_num = find_largest(a, b, c)
print(f"The largest number among {a}, {b}, and {c} is: {largest_num}")
```

```python
# 2. Write a program to check whether a given year is a leap year.

year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year.")
else:
    print(year, "is NOT a Leap Year.")
```

```python
# 3. Write a program to check if a character is a vowel or a consonant. 

char = input("Enter a letter: ")
if char in 'aeiouAEIOU':
    print(char, "is a vowel.")
else:
    print(char, "is a consonant.")
```

```python
# 4. Write a program to check whether a number is divisible by both 5 and 11. 

number = int(input("Enter a number: "))

if number % 5 == 0 and number % 11 == 0:
    print(number, "is divisible by both 5 and 11.")
else:
    print(number, "is NOT divisible by both 5 and 11.")
```

```python
# 5. Write a program to alculate the sum of first N natural numbers using a while loop. 

n = int(input("Enter a positive integer: "))

sum = 0
i = 1

while i <= n:
    sum += i
    i += 1

print("The sum of first", n, "natural numbers is:", sum)
```

```python
#6 Write a program to print the multiplication table of a given number using a while loop. 

num = int(input("Enter a number: "))

i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

```python
#7. Print a pyramid pattern:
'''* 
  *** 
 ***** 
*******'''

i = 1
while i <= 4:
    print(" " * (4 - i) + "*" * (2 * i - 1))
    i += 1
```

```python
# 8. Print: 

'''1 
12 
123 
1234 
12345'''

rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
